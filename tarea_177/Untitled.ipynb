```python
#HAGO UNA COPIA DE DF:
df2 = df.copy()

#ELIMINO LA COLUMNA DE NIF, DEBIDO A QUE NO APORTA VALOR O INFORMACIÓN A ANALIZAR:
df2 = df2.drop(columns=["NIF"])

#PARA CONOCER LAS PROPORCIONES DE LAS RESPUESTAS DE LAS COLUMNAS:
def proporciones(data):
    columns = ["Sexo", "Edad","Estatura", "Vacunadx", "Masa","Hijos", "Estado Civil", "Vacunadx", "CARNÉ CONDUCIR", "Consumidorx Alcohol Esporádico"   ]
    for col in columns:
        print(f"La columna {col} tiene las siguientes proporciones: ")
        print(dict(round(data[col].value_counts(normalize=True)*100,2)))
        
proporciones(df2)
```
